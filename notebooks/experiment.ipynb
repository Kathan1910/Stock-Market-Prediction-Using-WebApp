```python
import yfinance as yf
import pandas as pd
```

# 2. Define the Data Fetching Function

```python
def fetch_sensex_30_data(ticker_symbol, start_date, end_date):
    """
    Fetch historical stock data for a given ticker symbol from Yahoo Finance.
    
    Parameters:
    - ticker_symbol (str): The ticker symbol for the stock.
    - start_date (str): Start date in the format 'YYYY-MM-DD'.
    - end_date (str): End date in the format 'YYYY-MM-DD'.
    
    Returns:
    - DataFrame: A pandas DataFrame containing the historical stock data.
    """
    ticker_data = yf.Ticker(ticker_symbol)
    df = ticker_data.history(start=start_date, end=end_date)
    return df
```

# 3. Define the SENSEX 30 Ticker Symbols

```python
sensex_30_tickers = [
    "ASIANPAINT.BO", "AXISBANK.BO", "BAJAJ-AUTO.BO", "BAJFINANCE.BO", 
    "BAJAJFINSV.BO", "BHARTIARTL.BO", "DRREDDY.BO", "HCLTECH.BO", 
    "HDFC.BO", "HDFCBANK.BO", "HEROMOTOCO.BO", "HINDUNILVR.BO", 
    "ICICIBANK.BO", "INDUSINDBK.BO", "INFY.BO", "ITC.BO", 
    "KOTAKBANK.BO", "LT.BO", "M&M.BO", "MARUTI.BO", "NESTLEIND.BO", 
    "NTPC.BO", "ONGC.BO", "POWERGRID.BO", "RELIANCE.BO", "SBIN.BO", 
    "SUNPHARMA.BO", "TATAMOTORS.BO", "TATASTEEL.BO", "TCS.BO", "TECHM.BO"
]
```

# 4. Fetch Data for Each Ticker

```python
start_date = "2018-01-01"
end_date = "2023-10-01"

all_data = {}

for ticker in sensex_30_tickers:
    print(f"Fetching data for {ticker}...")
    data = fetch_sensex_30_data(ticker, start_date, end_date)
    all_data[ticker] = data
```

# Data PreProcessing

# 1. Handling Missing Values

```python
# Load the Data: 

import pandas as pd
import matplotlib.pyplot as plt

data_path = "data/raw/SENSEX/RELIANCE.BO.csv"  # Example for Reliance Industries Limited
df = pd.read_csv(data_path, parse_dates=['Date'], index_col='Date')
```

```python
# Visualize Missing Data:
# Plot the stock's closing prices to visually identify any gaps.

df['Close'].plot(figsize=(14, 7))
plt.title('Stock Closing Prices')
plt.show()
```

```python
# Identify Missing Dates:
# Check for missing dates in the dataset.

all_dates = pd.date_range(start=df.index.min(), end=df.index.max(), freq='B')
missing_dates = all_dates.difference(df.index)
print(missing_dates)
print(len(all_dates))
print(len(missing_dates))
```

```python
# Handle Missing Values:
# Depending on the number and nature of missing values, decide on a method to fill these gaps. 
# For this example, let's use forward fill.

df = df.reindex(all_dates)
df.fillna(method='ffill', inplace=True)
```

```python
# Visualize After Filling:
# Plot the stock's closing prices again to see the filled gaps.
df['Close'].plot(figsize=(14, 7))
plt.title('Stock Closing Prices After Filling Missing Values')
plt.show()
```

# 2. Handling Outliers

```python
# Visualize Data Distribution:
# Use a box plot to visualize the distribution of closing prices and identify potential outliers.

df['Close'].plot(kind='box', vert=False, figsize=(14, 6))
plt.title('Box Plot of Closing Prices')
plt.show()
```

```python
# Statistical Identification:
# Identify outliers using a method like the IQR (Interquartile Range).

Q1 = df['Close'].quantile(0.25)
Q3 = df['Close'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Close'] < lower_bound) | (df['Close'] > upper_bound)]
print(outliers)
```

```python
# Handle Outliers:
# Depending on the nature of the outliers, decide whether to remove them, cap them, or replace them. 
# For this example, let's cap them to the bounds.

df['Close'] = df['Close'].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Visualize After Handling Outliers:
# Plot the stock's closing prices again to see the changes.

df['Close'].plot(figsize=(14, 7))
plt.title('Stock Closing Prices After Handling Outliers')
plt.show()
```

# 3. Feature Engineering.

## 1. Technical Indicators

```python
# Moving Averages: 

# Simple Moving Average
df['SMA_30'] = df['Close'].rolling(window=30).mean()
df['SMA_60'] = df['Close'].rolling(window=60).mean()

# Exponential Moving Average
df['EMA_30'] = df['Close'].ewm(span=30, adjust=False).mean()
df['EMA_60'] = df['Close'].ewm(span=60, adjust=False).mean()

```

```python
# 1. RSI:

delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).fillna(0)
loss = (-delta.where(delta < 0, 0)).fillna(0)

avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()

rs = avg_gain / avg_loss
df['RSI'] = 100 - (100 / (1 + rs))
```

```python
## 2. MACD:

short_ema = df['Close'].ewm(span=12, adjust=False).mean()
long_ema = df['Close'].ewm(span=26, adjust=False).mean()

df['MACD'] = short_ema - long_ema
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
```

## 2. Lagged Features 

```python
# Creating 5 days lagged features
for i in range(1, 6):
    df[f'Lagged_{i}'] = df['Close'].shift(i)
```

```python
import seaborn as sns

correlation_matrix = df[['Close'] + [f'Lagged_{i}' for i in range(1, 6)]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Closing Price and Lagged Features')
plt.show()
```

```python
# This heatmap will give you an idea of how each lagged feature correlates with the current closing price. 
# Strong correlations indicate that the lagged feature might be a useful predictor for the current price.
```

## 3. Rolling Statistics

```python
df['Rolling_Mean_3'] = df['Close'].rolling(window=3).mean()
df['Rolling_Std_3'] = df['Close'].rolling(window=3).std()
```

```python
import pandas as pd

# Replace 'your_ticker_symbol.csv' with the desired file name
data_path = "data/processed/processed_ASIANPAINT.BO.csv"
data = pd.read_csv(data_path, parse_dates=['Date'], index_col='Date')

# Display the first few rows
data.head()
```

## 2. Train-Test Split:


```python
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]
```

## 3. Model Selection & Training:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Features and target
X_train = train.drop('Close', axis=1)
y_train = train['Close']
X_test = test.drop('Close', axis=1)
y_test = test['Close']

# Model training
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Evaluation
train_rmse = mean_squared_error(y_train, train_preds, squared=False)
test_rmse = mean_squared_error(y_test, test_preds, squared=False)

print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
```

## 4. Visualize Predictions:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
plt.plot(test.index, y_test, label='Actual Prices', color='blue')
plt.plot(test.index, test_preds, label='Predicted Prices', color='red', linestyle='--')
plt.title('Actual vs Predicted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

## Advanced Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predictions
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Evaluation metrics for training data
train_mae = mean_absolute_error(y_train, train_preds)
train_mse = mean_squared_error(y_train, train_preds)
train_rmse = mean_squared_error(y_train, train_preds, squared=False)
train_r2 = r2_score(y_train, train_preds)

# Evaluation metrics for testing data
test_mae = mean_absolute_error(y_test, test_preds)
test_mse = mean_squared_error(y_test, test_preds)
test_rmse = mean_squared_error(y_test, test_preds, squared=False)
test_r2 = r2_score(y_test, test_preds)

# Print the evaluation metrics
print("Training Data:")
print(f"MAE: {train_mae}")
print(f"MSE: {train_mse}")
print(f"RMSE: {train_rmse}")
print(f"R2 Score: {train_r2}")
print("\n")
print("Testing Data:")
print(f"MAE: {test_mae}")
print(f"MSE: {test_mse}")
print(f"RMSE: {test_rmse}")
print(f"R2 Score: {test_r2}")
```
